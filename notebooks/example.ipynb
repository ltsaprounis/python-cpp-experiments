```python
import numpy as np
from pycpp.my_linalg import dot_product, arma_dot_product, array_sum
```

```python
# Create test arrays
n = 1_000_000
a = np.random.rand(n)
b = np.random.rand(n)
print(f"Array size: {n:,}")
```

## Dot Product

```python
# Verify results match
cpp_dot = dot_product(a, b)
arma_dot = arma_dot_product(a, b)
np_dot = np.dot(a, b)
print(f"C++ dot_product:      {cpp_dot}")
print(f"C++ arma_dot_product: {arma_dot}")
print(f"NumPy np.dot:         {np_dot}")
```

```python
%%timeit
dot_product(a, b)
```

```python
%%timeit
arma_dot_product(a, b)
```

```python
%%timeit
np.dot(a, b)
```

## Array Sum

```python
# Verify results match
cpp_sum = array_sum(a, b)
np_sum = a + b
print(f"Results match: {np.allclose(cpp_sum, np_sum)}")
```

```python
%%timeit
array_sum(a, b)
```

```python
%%timeit
a + b
```
